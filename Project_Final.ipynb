### Import dependencies

```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame and explore data

```python
csv_file1 = "Barcelona1992.csv"
df1 = pd.read_csv(csv_file1)
df1.head()
```

### Create new data with select columns and rename columns

```python
csv_file1 = "Barcelona1992.csv"
df1 = pd.read_csv(csv_file1)
df1 = df1[['Pas 2','Pas 3','Col 1','Col 2', 'Col 3', 'Total']].copy()
df1 = df1.rename(columns={'Pas 2':'country', 'Pas 3':'country_id','Col 1':'gold_92','Col 2':'silver_92','Col 3':'bronze_92','Total':'total_92'})
csv_file2 = "Atlanta1996.csv"
df2 = pd.read_csv(csv_file2)
df2 = df2[['Pas 2','Pas 3','Col 1','Col 2', 'Col 3', 'Total']].copy()
df2 = df2.rename(columns={'Pas 2':'country', 'Pas 3':'country_id','Col 1':'gold_96','Col 2':'silver_96','Col 3':'bronze_96','Total':'total_96'})
csv_file3 = "Sidney2000.csv"
df3 = pd.read_csv(csv_file3)
df3 = df3[['Pas 2','Pas 3','Col 1','Col 2', 'Col 3', 'Total']].copy()
df3 = df3.rename(columns={'Pas 2':'country', 'Pas 3':'country_id','Col 1':'gold_00','Col 2':'silver_00','Col 3':'bronze_00','Total':'total_00'})
csv_file4 = "Atenas2004.csv"
df4 = pd.read_csv(csv_file4)
df4 = df4[['Pas 2','Pas 3','Col 1','Col 2', 'Col 3', 'Total']].copy()
df4 = df4.rename(columns={'Pas 2':'country', 'Pas 3':'country_id','Col 1':'gold_04','Col 2':'silver_04','Col 3':'bronze_04','Total':'total_04'})
csv_file5 = "Pekin2008.csv"
df5 = pd.read_csv(csv_file5)
df5 = df5[['Pas 2','Pas 3','Col 1','Col 2', 'Col 3', 'Total']].copy()
df5 = df5.rename(columns={'Pas 2':'country', 'Pas 3':'country_id','Col 1':'gold_08','Col 2':'silver_08','Col 3':'bronze_08','Total':'total_08'})
csv_file6 = "Londres2012.csv"
df6 = pd.read_csv(csv_file6)
df6 = df6[['Pas 2','Pas 3','Col 1','Col 2', 'Col 3', 'Total']].copy()
df6 = df6.rename(columns={'Pas 2':'country', 'Pas 3':'country_id','Col 1':'gold_12','Col 2':'silver_12','Col 3':'bronze_12','Total':'total_12'})
csv_file7 = "Rio2016.csv"
df7 = pd.read_csv(csv_file7)
df7 = df7[['Pas 2','Pas 3','Col 1','Col 2', 'Col 3', 'Total']].copy()
df7 = df7.rename(columns={'Pas 2':'country', 'Pas 3':'country_id','Col 1':'gold_16','Col 2':'silver_16','Col 3':'bronze_16','Total':'total_16'})
```

```python
df1.head()
```

### Connect to local data base in postgres

```python
rds_connection_string = "postgres:postgres@localhost:5432/OlympicGames_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
df1.to_sql(name='barcelona_1992', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from barcelona_1992', con=engine).head()
```

```python
df2.to_sql(name='atlanta_1996', con=engine, if_exists='append', index=False)
df3.to_sql(name='sidney_2000', con=engine, if_exists='append', index=False)
df4.to_sql(name='atenas_2004', con=engine, if_exists='append', index=False)
df5.to_sql(name='pekin_2008', con=engine, if_exists='append', index=False)
df6.to_sql(name='londres_2012', con=engine, if_exists='append', index=False)
df7.to_sql(name='rio_2016', con=engine, if_exists='append', index=False)
```

```python
df8 = pd.merge(df1,df2, on='country',how='inner')
```

```python
df8.head()
```

```python
df9 = pd.merge(df8,df3,on='country',how='inner')
df9.head()
```

```python
df10 = pd.merge(df9,df4,on='country',how='inner')
df10.head()
```

```python
df11 = pd.merge(df10,df5,on='country',how='inner')
df11.head()
```

```python
df12 = pd.merge(df11,df6,on='country',how='inner')
df12.head()
```

```python
final_df = pd.merge(df12,df7,on='country',how='inner')
final_df.head()
```

```python
cons_df = final_df[['country','country_id','gold_92','silver_92','bronze_92','gold_96','silver_96','bronze_96','gold_00','silver_00','bronze_00','gold_04','silver_04','bronze_04','gold_08','silver_08','bronze_08','gold_12','silver_12','bronze_12','gold_16','silver_16','bronze_16','total_92','total_96','total_00','total_04','total_08','total_12','total_16']]
cons_df.head()
```

```python
gold_df = cons_df[['country','country_id','gold_92','gold_96','gold_00','gold_04','gold_08','gold_12','gold_16']]
gold_df['total_gold']=gold_df['gold_92']+gold_df['gold_96']+gold_df['gold_00']+gold_df['gold_04']+gold_df['gold_08']+gold_df['gold_12']+gold_df['gold_16']
gold_df.head()
```

```python
silver_df = cons_df[['country','country_id','silver_92','silver_96','silver_00','silver_04','silver_08','silver_12','silver_16']]
silver_df['total_silver']=silver_df['silver_92']+silver_df['silver_96']+silver_df['silver_00']+silver_df['silver_04']+silver_df['silver_08']+silver_df['silver_12']+silver_df['silver_16']
silver_df.head()
```

```python
bronze_df = cons_df[['country','country_id','bronze_92','bronze_96','bronze_00','bronze_04','bronze_08','bronze_12','bronze_16']]
bronze_df['total_bronze']=bronze_df['bronze_92']+bronze_df['bronze_96']+bronze_df['bronze_00']+bronze_df['bronze_04']+bronze_df['bronze_08']+bronze_df['bronze_12']+bronze_df['bronze_16']
bronze_df.head()
```

```python
gold_df.to_sql(name='gold', con=engine, if_exists='append', index=False)
silver_df.to_sql(name='silver', con=engine, if_exists='append', index=False)
bronze_df.to_sql(name='bronze', con=engine, if_exists='append', index=False)
```

```python
gold_df.head()
```

# Create the Great Total table

```python
total_df = cons_df[['country','country_id']]
total_df['total_gold'] = gold_df['total_gold']
total_df['total_silver'] = silver_df['total_silver']
total_df['total_bronze'] = bronze_df['total_bronze']
total_df['total_total'] = bronze_df['total_bronze'] + silver_df['total_silver'] + gold_df['total_gold']
total_df.sort_values('total_total', ascending = False)
 
total_df.head()
```

# Load DataFrame to Posgres

```python
total_df.to_sql(name='greattotal', con=engine, if_exists='append', index=False)
```
